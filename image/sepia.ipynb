Using python, I would give the picture sepia effect, a common Instagram filter.

### Background

Each pixel in the picture is composed out of 3 dimension vector:$$\begin{bmatrix}
R\\
G\\
B
\end{bmatrix}$$to get the sepia filter, I would transform each vector into a new vector using the sepia matrix: 
$$T=\begin{bmatrix}
0.393 &	0.769	& 0.189\\
0.349	&0.686& 	0.168\\
0.272	&0.534&	0.131\\
\end{bmatrix}$$


```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import os
```

First, I would convert the chosen image into a matrix array

```python
image = Image.open('./IMG_5858.jpg').convert('RGB')
img = np.array(image)
plt.imshow(img)
plt.show()
```

Then, I would set the sepia matrix, I would loop over each row in array, and transfer each pixel vector using the sepia matrix.

```python
sepiaMatrix = np.matrix([[ 0.393, 0.769, 0.189],
                      [ 0.349, 0.686, 0.168],
                      [ 0.272, 0.534, 0.131]                                  
    ])

filt = np.array([x * sepiaMatrix.T for x in img] )
```

Since highest value of that each row in the pixel vector is 255, I would change each item that is above 255 into 255.

```python
filt[np.where(filt>255)] = 255
```

 The last step, I would convert the array back into an jpeg

```python
t = Image.fromarray(filt.astype('uint8'))
plt.imshow(t)
plt.show()
```

## Linear Algebra Concepts
The conecpt that have been used here were:
1. Change of basis: In order to see the picture in sepia filter, I change the pixal basis to the basis of the sepia matrix.
2. Linear Transformation: I did the change of basis via linear transformation.
3. Matrix Multiplication: In order to do the transformation, I had to do matrix multiplcation between the matrix and the vector.
